```python
import pandas as pd
import statistics
from scipy import stats
import seaborn as sns  # Librera para graficar la funcin de densidad
import matplotlib.pyplot as plt

# Leer el archivo CSV
df = pd.read_csv("D:/docsUNI/9 Semestre UNI/Teoria de la informacion/Estadstica descriptiva/data/dataset_thyroid_sick.csv")
```

```python
# Seleccionar la columna de edad
column_age = df["age"]

# Convertir la columna a tipo numrico, eliminando o ignorando valores no numricos
column_age = pd.to_numeric(column_age, errors='coerce')

# Eliminar valores NaN
age_data = column_age.dropna().tolist()
```

```python
print (df.head())
print (df.info())
```

```python
# Calcular las estadsticas usando la librera statistics y scipy
try:
    moda = statistics.mode(age_data)
except statistics.StatisticsError:
    moda = 'Sin moda'

media = statistics.mean(age_data)
mediana = statistics.median(age_data)
varianza = stats.tvar(age_data)

# Imprimir los resultados
print("Moda:", moda)
print("Media:", media)
print("Mediana:", mediana)
print("Varianza:", varianza)
```

```python
# Graficar la funcin de densidad de la columna de edad
plt.figure(figsize=(10, 6))
sns.kdeplot(age_data, shade=True, color="blue")
plt.title("Funcin de Densidad de la Edad")
plt.xlabel("Edad")
plt.ylabel("Densidad")
plt.show()
```
